```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
# Import data
df = pd.read_csv('../Resources/Cleaning2/merged_events_bio.csv')
df.head(5)
```

```python
df = df.rename(columns={'Sex': 'Gender'})
df.head(5)
```

```python
# create a function to update the plot based on the selected year
def update_plot(year):
    # filter by year and gender
    filtered_df = df[(df['Year'] == year) * ([df['Gender'] == gender]) & ([df['Sport'] == sport]) & ([df['Country_Name'] == country])]
  
    # filter out rows where no medal was received
    filtered_df = filtered_df[~(filtered_df['Medal'] == 'na')]
    
    # create a pivot table to count medals by age, gender, and medal type
    medal_counts = pd.pivot_table(filtered_df, index=['Age', 'Gender', 'Sport', 'Country_Name'], columns='Medal', values='Name', aggfunc='count', fill_value=0)
  
    # create a stacked bar chart
    medal_counts.plot(kind='bar', stacked=True)

    # set x and y axis labels
    plt.xlabel('Age and Gender')
    plt.ylabel('Medal Count')

    # show the plot
    plt.show()
```

```python
# create dropdowns to select data
years = df['Year'].unique()
genders = df['Gender'].unique()
sports = df['Sport'].unique()
countries = df['Country_Name'].unique()
interact(update_plot, year=years, gender=genders, sport=sports, country=countries);
```
