```python
# Insert all countries that participated in the Olympics
country = pd.read_csv('../Resources/Olympics_Country.csv')
country = country.rename(columns={"country": "Country_Name", "country_noc": "Country_Code"})
print(country.shape)
print(country.head())
```

```python
# Insert coordinates for participating Olympics
coordinates = pd.read_csv('country_coordinate.csv')
coordinates = coordinates.rename(columns={"name": "Country_Name"})
coordinates.head()
```

```python
# Merge coordinates
country_coordinates = pd.merge(country, coordinates, on ="Country_Name")
country_coordinates = country_coordinates.reset_index()
country_coordinates = country_coordinates.drop(['index', 'country'], axis=1)
country_coordinates
```
