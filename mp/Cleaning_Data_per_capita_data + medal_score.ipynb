## Import Tables

```python
import pandas as pd
import numpy as np
```

```python
gdp = pd.read_csv('../Resources/gdp_pop_medals/gpd_medals_total.csv')
gdp.tail()
```

```python
pop = pd.read_csv('../Resources/gdp_pop_medals/pop_medals_total.csv')
pop.tail()
```

```python
population = pop[["Country_Name", "Year", "Population"]]
population.tail()
```

```python
df = pd.merge(gdp, population, on= ["Country_Name", "Year"]).reset_index()
df.head()
```

```python
df = df.drop('index', axis=1)
df= df[["Country_Name", "Country_Code", "Year", "GDP", "Population", "Gold", "Silver", "Bronze", "Total_Medals", "Season"]]
df
```

```python
# Reset GDP for capita calculation
df['GDP'] = df['GDP'].multiply(1000)
df['GDP'] = df['GDP'].fillna(0)
df
```

```python
df['Population'] = df['Population'].fillna(0)
```

```python
#calculate GDP per Capita
df['GDP_per_capita'] = df['GDP'] / df['Population']
df
```

```python
# multiply medal by points
df["Gold_Score"] = df["Gold"]*3
df["Silver_Score"] = df["Silver"]*2
df["Bronze_Score"] = df["Bronze"]*1
df['Total_Score'] = df['Gold_Score'] + df['Silver_Score'] + df['Bronze_Score']
df
```

```python
# calculate medal count per GPD
df['Medal_score_per_GDP'] = df['Total_Score'] / df['GDP']
df
```

```python
# calculate medal count per GPD
df['Medal_score_per_capita'] = df['Total_Score'] / df['Population']
df
```

```python
# replace inf with zero
df['Medal_score_per_GDP'] = df['Medal_score_per_GDP'].replace([np.inf, -np.inf], 0)
df.head()
```

```python
df.columns
```

```python
df = df[['Country_Name', 'Country_Code','Season', 'Year', 'GDP', 'Population', 'GDP_per_capita', 'Gold',
       'Silver', 'Bronze', 'Total_Medals', 'Gold_Score',
       'Silver_Score', 'Bronze_Score', 'Total_Score', 'Medal_score_per_GDP','Medal_score_per_capita']]
df
```

```python
zero_gdp_pop_rows = df[(df['GDP'] == 0) | (df['Population'] == 0)]
zero_gdp_pop_rows
```

```python
zero_gdp_pop_rows.to_csv('GDP_missing_data.csv')
df.to_csv('calculations_per_capita_and_medal_score.csv')
```
