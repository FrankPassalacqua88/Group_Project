```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
# Insert data
if 'data' not in pn.state.cache.keys():


    df = pd.read_csv('olympic_data_country_medal_year.csv')
else: 

    df = pn.state.cache['data']
```

```python
df
```

```python
# Insert coordinates for participating Olympics
coordinates = pd.read_csv('Resources/country_coordinate.csv')
coordinates = coordinates.rename(columns={"name": "Country_Name"})
coordinates.head()
```

```python
# Merge coordinates
df = pd.merge(df, coordinates, on ="Country_Name")
df = df.reset_index()
df= df.sort_values("Year", ascending = True)
df.tail()
```

```python
# Version 1 does not display data in format I would like but just testing this out
import pandas as pd

import plotly.express as px

fig = px.scatter_mapbox(df, lat="latitude", lon="longitude", hover_name="Country_Name", hover_data=["GDP", "Y_Total"],
                       color_discrete_sequence=["fuchsia"], zoom=3, height=300,
                       title="Olymic Countries by GDP and Population")
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
import plotly.express as px
df= df.sort_values("Year", ascending = True)
fig1 = px.choropleth(df, 
                    locations="Country_Code",
                    color="GDP_Per_Capita",
                    hover_name="Country_Name",
                    animation_frame="Year",
                    hover_data=["GDP_Per_Capita", "A_Total"],
                    range_color=[0,60000],
                    color_continuous_scale=px.colors.sequential.Plasma_r)

fig1.update_xaxes(title_text='Top Medal Score')
fig1.show()
```

```python
# Radio buttons for Population, GDP, GDP Per Capita
demographics = pn.widgets.RadioButtonGroup(
    name='demographic', 
    options=['Population', 'GDP', 'GDP Per Capita',],
    button_type='success'
)
```
