```python
import pandas as pd
```

```python
events_df = pd.read_csv('../Resources/Olympic_Athlete_Event_Results_1896_2022.csv')
events_df.head()
```

```python
events_df = events_df.rename(columns={"country_noc": "Country_Code", 'athlete': "Name", "sport":"Sport", "medal": "Medal", "event": "Sex"})
events_df = events_df.drop(['edition_id', 'result_id', 'athlete_id', 'pos'], axis =1)
events_df.head()
```

```python
events_df['Sex'] = events_df['Sex'].replace('Men', 'M', inplace=True)
events_df['Sex'] = events_df['Sex'].replace('Women', 'F',inplace=True)
```

```python
events_df[['Year','Season', 'delete']] = events_df.edition.str.split(expand=True)
events_df = events_df.drop(['delete', 'edition'], axis = 1)
events_df.head()
```

```python
events_df.to_csv('../Resources/clean_events.csv')
```
