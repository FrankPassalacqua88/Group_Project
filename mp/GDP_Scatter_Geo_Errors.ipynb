```python
import pandas as pd
import numpy as np
```

```python
# Insert data
df = pd.read_csv('olympic_data_country_medal_year.csv')
df.tail()
```

```python
# Insert coordinates for participating Olympics
coordinates = pd.read_csv('Resources/country_coordinate.csv')
coordinates = coordinates.rename(columns={"name": "Country_Name"})
coordinates.head()
```

```python
# Merge coordinates
df = pd.merge(df, coordinates, on ="Country_Name")
df = df.reset_index()
df.head()
```

```python
df = df.drop(["Unnamed: 0", "index"], axis=1)
```

```python
df = df.sort_values(['Year','Country_Name'], ascending=True)
```

```python
import plotly.express as px
fig = px.scatter_geo(df, locations="Country_Code", color="GDP",
                     hover_name="Country_Name", size=("A_Total"),
                     animation_frame="Year",
                     projection="natural earth")
            
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px

fig = make_subplots(rows=2, cols=2)

# Add first chart to subplot 1
fig.add_trace(go.Bar(
    y=df['Country_Name'],
    x=df['GDP'],
    name='GDP',
    marker_color='lightsalmon',
    orientation='h'
), row=1, col=1)
fig.update_xaxes(title_text='Top GDP', row=1, col=1)

fig.add_trace(go.Bar(
    y=df['Country_Name'],
    x=df['A_Total'],
    name='Net Total',
    marker_color='indianred',
    orientation='h'
), row=1, col=2)
fig.update_xaxes(title_text='Top Medal Count', row=1, col=2)


fig.update_layout(height=800, width=1000, title_text="Olympic Data")

fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px

# create an empty subplot
fig = make_subplots(rows=2, cols=1, specs=[[{"type": "bar"}], [{"type": "choropleth"}]])

# add the choropleth trace
fig.add_trace(
    px.choropleth(
        df, 
        locations="Country_Code",
        color="GDP2",
        hover_name="Country_Name",
        color_continuous_scale=px.colors.sequential.Plasma
    ).data[0], 
    row=2, col=1
)

# create a bar chart for each year
for year in df['Year'].unique():
    df_year = df[df['Year']==year].sort_values('GDP', ascending=False).head(10)
    fig.add_trace(
        go.Bar(
            x=df_year['GDP'],
            y=df_year['Country_Name'],
            name=str(year),
            marker_color='lightsalmon'
        ),
        row=1, col=1
    )

# update layout
fig.update_xaxes(title_text='Top GDP', row=1, col=1)
fig.update_xaxes(title_text='Year', range=[df['Year'].min(), df['Year'].max()], row=2, col=1)
fig.update_yaxes(title_text='Country', row=1, col=1)
fig.update_yaxes(title_text='GDP', row=1, col=1)
fig.update_layout(height=800, title_text='Top 10 Countries by GDP (1960-2019)')

# add animation
fig.update_layout(updatemenus=[dict(
    type='buttons',
    showactive=False,
    buttons=[dict(
        label='Play',
        method='animate',
        args=[None, dict(frame=dict(duration=5, redraw=True), fromcurrent=True)]
    ), dict(
        label='Pause',
        method='animate',
        args=[[None], dict(frame=dict(duration=0, redraw=True), mode='immediate', transition=dict(duration=0))]
    )]
)])

# show figure
fig.show()
```
