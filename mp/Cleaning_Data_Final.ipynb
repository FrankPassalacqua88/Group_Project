```python
import pandas as pd
import numpy as np
```

```python
gdp = pd.read_csv('../Resources/gdp_pop_medals/gpd_medals_total.csv')
```

```python
gdp2 = pd.read_csv('../Resources/Original/gdp_1960_2021.csv')
pop2 = pd.read_csv('../Resources/Original/population_1960_2021.csv')
gdp2.head()
```

```python
gdp2 = gdp2.rename(columns = {'Country Name':'Country_Name',
                            'Country Code':'Country_Code',
                            'Indicator Name':'Indicator_Name',
                           'Indicator Code':'Indicator_Code'})
gdp2 = gdp2.drop(['Indicator_Name','Indicator_Code'], axis = 1, inplace = False)
pop2 = pop2.rename(columns = {'Country Name':'Country_Name',
                            'Country Code':'Country_Code',
                            'Indicator Name':'Indicator_Name',
                           'Indicator Code':'Indicator_Code'})
pop2 = pop2.drop(['Indicator_Name','Indicator_Code'], axis = 1, inplace = False)
pop2.head()
```

```python
# Melt the GDP dataframe
gdp2 = pd.melt(gdp2,id_vars =['Country_Name','Country_Code'], var_name = 'Year', value_name = 'GDP_Per_Capita')
gdp2['Year'] = gdp2['Year'].astype('int64')
gdp2.dtypes
pop2 = pd.melt(pop2,id_vars =['Country_Name','Country_Code'], var_name = 'Year', value_name = 'Pop2')
pop2['Year'] = pop2['Year'].astype('int64')
pop2.dtypes
```

```python
# Reset GDP for capita calculation
df['GDP_Per_Capita'] = df['GDP_Per_Capita'].fillna(0)
df['Gold'] = df['Gold'].fillna(0)
df['Silver'] = df['Silver'].fillna(0)
df['Bronze'] = df['Bronze'].fillna(0)
df['Total_Medals'] = df['Total_Medals'].fillna(0)
df.head()
```

```python
#calculate GDP
df['GDP'] = df['GDP_Per_Capita'] * df['Pop2']
df.head()
```

```python
df.columns
```

```python
df = df.sort_values(['Year','Country_Name'], ascending=True)
df.head()
```

```python
df[['A_Total', 'A_Gold', 'A_Silver', 'A_Bronze']] = df.groupby(['Country_Name'])[['Total_Medals','Gold', 'Silver', 'Bronze']].cumsum()
df[df['Country_Name'] == 'United States'].tail()
```

```python
df
```

```python
df.columns
```

```python
df = df.rename(columns = {'Country_Code_x':'Country_Code', 'Season_x' :'Season'})
```

```python
# Y_value is for current year, A_value is accumulated up to current year

df = df.rename(columns={'Pop2': 'Population', 'Gold': 'Y_Gold', 'Silver':'Y_Silver', 'Bronze': 'Y_Bronze', 'Total_Medals': 'Y_Total'})
df = df[['Country_Name', 'Country_Code','Year','Season', 'Population', 'GDP', 'GDP_Per_Capita','Y_Gold', 'Y_Silver', 'Y_Bronze', 'Y_Total', 'A_Gold', 'A_Silver', 'A_Bronze','A_Total']]

df
```

```python
df['Y_Capita_Per_Medal'] = df['Y_Total']/df['Population']
df['A_Capita_Per_Medal'] = df['A_Total']/df['Population']
df['Y_Capita_Per_GDP'] = df['Y_Total']/df['GDP']
df['A_Capita_Per_GDP'] = df['A_Total']/df['GDP']
df['A_Score'] = df['A_Gold']*3 + df['A_Silver']*2 + df['A_Bronze']
df['Y_Score'] = df['Y_Gold']*3 + df['Y_Silver']*2 + df['Y_Bronze']
```

```python
df['Y_Score_Per_Capita'] = df['Population']/df['Y_Score']
df['A_Score_Per_Capita'] = df['Population']/df['A_Score']
```

```python
df.columns
```

```python
df = df[['Country_Name', 'Country_Code', 'Year', 'Season', 'Population', 'GDP',
       'GDP_Per_Capita', 'Y_Gold', 'Y_Silver', 'Y_Bronze', 'Y_Total', 'A_Gold',
       'A_Silver', 'A_Bronze', 'A_Total', 'Y_Capita_Per_Medal',
       'A_Capita_Per_Medal', 'Y_Capita_Per_GDP', 'A_Capita_Per_GDP', 'A_Score',
       'Y_Score', 'Y_Score_Per_Capita', 'A_Score_Per_Capita']]
```

```python
zero_gdp_pop_rows = df[(df['GDP'] == 0) | (df['Population'] == 0)]
zero_gdp_pop_rows.tail()
```

```python
df.loc[df['Season'].isna() & df['Year'].isin([2020, 2016, 2012, 2008, 2004, 2000, 1996, 1992, 1988,1984,1980,1976,1972,1968,1964]), 'Season'] = 'Summer'
df.loc[df['Season'].isna() & df['Year'].isin([2018, 2014, 2010, 2006, 2002, 1998, 1994, 1990,1986,1982,1978,1974,1970,1966,1962]), 'Season'] = 'Winter'
df[df['Season'].isna()]
```

```python
df = df.replace([float('inf'), float('-inf')], 0)
df
```
