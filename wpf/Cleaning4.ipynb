```python
tally = tally.drop(['Country_Code','Season', 'Gold','Silver','Bronze'],axis = 1)
tally['Country_Name'] = tally['Country_Name'].replace({'Republic of Korea': 'South Korea'})
tally['Country_Name'] = tally['Country_Name'].replace({"Democratic People's Republic of Korea":'North Korea'})
tally['Country_Name'] = tally['Country_Name'].replace({'Soviet Union':'Russian Federation'})
tally['Country_Name'] = tally['Country_Name'].replace({'East Germany':'Germany'})
tally['Country_Name'] = tally['Country_Name'].replace({'West Germany':'Germany'})

tally
```

```python
# Pivoting table, replacing NAs with 0, preventing float dtype by adding int, resetting index
tally = tally.pivot_table('Total_Medals', ['Country_Name'], 'Year').fillna(0).astype(int).reset_index()

tally
```

```python
# Looping through the columns and adding the previous column to each column
add_year = tally.columns[1]

for year in tally.columns[2:]:

    tally[year] = tally[add_year] + tally[year]

    add_year = year
```

```python
tally.loc[tally["Country_Name"]=='Russian Federation']
```

```python
tally.to_excel('pivot_medals.xlsx')
```

```python
tally3 = tally.drop(['Country_Code', 'Season'], axis = 1)
tally3 = tally3.sort_values(by = ['Year'], ascending = True)
tally3['Cumulative'] = tally3.groupby(['Country_Name'])['Total_Medals'].cumsum()
tally3['Gold_Cumulative'] = tally3.groupby(['Country_Name'])['Gold'].cumsum()
tally3['Silver_Cumulative'] = tally3.groupby(['Country_Name'])['Silver'].cumsum()
tally3['Bronze_Cumulative'] = tally3.groupby(['Country_Name'])['Bronze'].cumsum()

tally3.head()
```

```python
tally3['Year'] = tally3.Year.astype('int32')
tally3['Year'] = pd.to_datetime(tally3.Year, format='%Y')
tally3.dtypes
```

```python
summer_tally = tally3.loc[tally3['Season'] == 'Summer']
summer_tally['Cumulative'] = summer_tally.groupby(['Country_Name'])['Total_Medals'].cumsum()
summer_tally.head()
```

```python
winter_tally = tally3.loc[tally3['Season'] == 'Winter']
winter_tally['Cumulative'] = winter_tally.groupby(['Country_Name'])['Total_Medals'].cumsum()
winter_tally.head()
```

```python
tally3.to_csv('cumulative_medal.csv')
tally3.to_excel('cumulative_medals.xls')
summer_tally.to_csv('summer_cumulative.csv')

winter_tally.to_csv('winter_cumulative.csv')
```
