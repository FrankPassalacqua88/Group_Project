```python
# Reading in csv
medals_df = pd.read_csv('../Resources/old/olympics_medals_country_wise.csv')

medals_df.head()
```

```python
# Checking columns for unneeded characters
medals_df.columns
```

```python
# Renaming columns
medals_df = medals_df.rename(columns={"ioc_code ": "Country Code", "countries ": "Country Name", "total_total ": 'total_total'})
# Further renaming columns for parity
medals_df = medals_df.rename(columns = {'Country Name': 'Country_Name', 
                                        'Country Code' : 'Country_Code', 
                                        'summer_participations': 'Summer_Participation',
                                       'summer_gold': 'Summer_Gold',
                                       'summer_silver':'Summer_Silver',
                                       'summer_bronze':'Summer_Bronze',
                                       'summer_total':'Summer_Total',
                                       'winter_participations':'Winter_Participation',
                                       'winter_gold':'Winter_Gold',
                                       'winter_silver':'Winter_Silver',
                                       'winter_bronze':'Winter_Bronze',
                                       'winter_total':'Winter_Total',
                                       'total_participation':'Total_Participation',
                                       'total_gold':'Total_Gold',
                                       'total_silver':'Total_Silver',
                                       'total_bronze':'Total_Bronze',
                                       'total_total':'Total_Medals'})
medals_df.columns
```

```python
# Removing '()' in Country_Code like Melinda
medals_df['Country_Code'] = medals_df['Country_Code'].str.replace(r'\(|\)', '')
medals_df.head()
```

```python
medals_df['Country_Name'].unique()
```

```python
medals_df.dtypes
```

```python
# Changed Summer_Gold column from object to int64
medals_df['Summer_Gold'] = medals_df['Summer_Gold'].replace("1,060","1060")
medals_df['Summer_Gold'] = medals_df['Summer_Gold'].astype('int64')
medals_df.dtypes
```

```python
#Changed Summer_Total column from object to int64
medals_df['Summer_Total'] = medals_df['Summer_Total'].replace('1,010','1010').replace('2,629','2629')
medals_df['Summer_Total'] = medals_df['Summer_Total'].astype('int64')
medals_df.dtypes
```

```python
# Changed Total_Gold column from object to int64
medals_df['Total_Gold'] = medals_df['Total_Gold'].replace('1,173','1173')
medals_df['Total_Gold'] = medals_df['Total_Gold'].astype('int64')
medals_df.dtypes
```

```python
# Changed Total_Medals column from object to int64
medals_df['Total_Medals'] = medals_df['Total_Medals'].replace('1,204','1204').replace('2,959','2959')
medals_df['Total_Medals'] = medals_df['Total_Medals'].astype('int64')
medals_df.dtypes
```

```python
# Medals DF Country_Names seems more clean than GDP and population DFs. 
# Should rows in GDP/Population be dropped to match Medals Country_Names? 
```

```python
# Making csv file for cleaned up data (temporary)
medals_df.to_csv('clean_medals.csv', index = False, header = True)
```
