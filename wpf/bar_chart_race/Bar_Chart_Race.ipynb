```python
import pandas as pd

import bar_chart_race as bcr

import panel as pn

pn.extension()
```

```python
df = pd.read_csv('../../Resources/Cleaning2/clean_medal_tally.csv')

df.head()
```

```python
df = df.drop(['Country_Code', 'Gold', 'Silver','Bronze'], axis = 1)

summer = df.loc[df['Season'] == 'Summer']

winter = df.loc[df['Season'] == 'Winter']

```

```python
summer = summer.drop(['Season'], axis = 1)
winter = winter.drop(['Season'], axis = 1)

summer['Country_Name'] = summer['Country_Name'].replace({'ROC':'Russian Federation'})
winter['Country_Name'] = winter['Country_Name'].replace({'ROC':'Russian Federation'})
```

```python
winter['Cumulative'] = winter.groupby(['Country_Name'])['Total_Medals'].cumsum()

summer['Cumulative'] = summer.groupby(['Country_Name'])['Total_Medals'].cumsum()

summer = summer.drop(['Total_Medals'], axis = 1)

winter = winter.drop(['Total_Medals'], axis = 1)
```

```python
summer = summer.pivot_table('Cumulative', ['Year'], 'Country_Name').fillna(0).astype(int).reset_index()

```

```python
winter = winter.pivot_table('Cumulative', ['Year'], 'Country_Name').fillna(0).astype(int).reset_index()

```

```python
summer = summer[['Year','Afghanistan','Greece','United States','Germany','France',
                 'Great Britain','Denmark','Hungary','Austria','Switzerland',"People's Republic of China",
                'Australia','Belgium','Mixed team','Canada','Romania','Cuba','Algeria','Finland','Norway','Italy',
                'Russian Federation','Soviet Union','Japan','East Germany','Poland']]
```

```python
winter = winter[['Year','Norway','Finland','Great Britain','United States','Austria',
                 'France','Switzerland','Sweden','Belgium','Canada','Czechoslovakia',
                'Hungary','Germany','Soviet Union','Netherlands','East Germany','Russian Federation']]
```

```python
summer.to_csv('summer_pivot.csv', index = False)
winter.to_csv('winter_pivot.csv', index = False)
```
