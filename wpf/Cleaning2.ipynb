```python
import pandas as pd
```

```python
bio = pd.read_csv('../Resources/Olympic_Athlete_Bio.csv')

bio.head()
```

```python
# Checking columns for unneeded characters
bio.columns
```

```python
# Renaming columns for parity

bio = bio.rename(columns = {'athlete_id':'Athlete_ID', 'name':'Name','sex':'Sex',
                 'born':'Birthyear','height':'Height(cm)','weight':'Weight(kg)',
                'country':'Country_Name','country_noc':'Country_Code',
                 'description':'Description','special_notes':'Special_Notes'})

bio.columns
```

```python
bio['Sex'] = bio['Sex'].replace({'Male':'M','Female':'F'})

bio.head()
```

```python
# Dropping unecessary columns

bio = bio.drop(['Description','Special_Notes'], axis = 1)

bio.head()
```

```python
bio.dtypes
```

```python
bio["Birthyear"] = pd.to_datetime(bio['Birthyear'], errors = 'coerce')
#bio.loc[(bio['Birthyear'] == 'na')]

bio.dtypes
```

```python
bio['column'] = bio['Birthyear'].str.split('-')

bio.head()
```

```python
bio_models = pd.DataFrame(bio['column'].tolist()).fillna('').add_prefix('model_')

bio_models = bio_models.drop(['model_1','model_2'], axis = 1)

bio_models.head()
```

```python
new_bio = pd.concat([bio,bio_models], axis = 1)

new_bio.head()
```

```python
# Reading in csv
event = pd.read_csv('../Resources/Olympic_Athlete_Event_Results_1896_2022.csv')

event.head()
```

```python
# Checking columns for unneeded characters

event.columns
```

```python
# Splitting edition into list

event['column'] = event['edition'].str.split(' ')

event.head()
```

```python
# Making new DataFrame to split 
event_models = pd.DataFrame(event['column'].tolist()).fillna('').add_prefix('model_')

event_models = event_models.drop(['model_2','model_3'], axis = 1)

event_models.head()
```

```python
# Concatinating DFs, dropping columns, renaming columns
new_event = pd.concat([event,event_models], axis = 1)

new_event = new_event.drop(['edition_id','edition','column', 'pos'],axis = 1)

new_event = new_event.rename(columns = {'country_noc':'Country_Code',
                                        'model_0':'Year','model_1':'Season','sport':'Sport','event':'Event',
                                       'result_id':'Result_ID','athlete':'Name','athlete_id':'Athlete_ID','medal':'Medal'})

new_event.head()
```

```python
# Merging the new_bio DF and the new_event DF and making new column Age to display age of athletes

merge_df = pd.merge(new_bio,new_event, on = 'Athlete_ID')

merge_df['model_0'] = pd.to_numeric(merge_df['model_0'], errors = 'coerce')

merge_df['Year'] = merge_df['Year'].astype('int64', errors = 'ignore')

merge_df['Age'] = merge_df['Year'] - merge_df['model_0'] 

merge_df['Age'] = merge_df['Age'].astype('Int64')
merge_df.head()
```

```python
# Dropping, renaming, and reorganizing DF

merge_df = merge_df.drop(['column','model_0','Country_Code_y','Name_y','Birthyear', 'Event'], axis = 1)

merge_df = merge_df.rename(columns = {'Name_x':'Name','Country_Code_x':'Country_Code'})

merge_df = merge_df[['Athlete_ID','Name', 'Age', 'Sex', 'Height(cm)', 'Weight(kg)','Sport',
               'Country_Name','Country_Code','Year','Season','isTeamSport','Medal','Result_ID']]

merge_df.head()
```

```python
# Changing dtypes of Height and Weight columns to numeric
merge_df['Height(cm)'] = pd.to_numeric(merge_df['Height(cm)'], errors = 'coerce')

merge_df['Weight(kg)'] = pd.to_numeric(merge_df['Weight(kg)'], errors = 'coerce')

merge_df.dtypes
```

```python
# Making csv file (initial)

merge_df.to_csv('merged_events_bio.csv', index = False, header = True)
```

```python
# Reading in csv file
results = pd.read_csv('../Resources/Olympic_Results.csv')

results.head()
```

```python
# Checking columns for unneeded characters

results.columns
```

```python
# Renaming columns for parity

results = results.rename(columns = {'result_id':'Result_ID','event_title':'Event_Title','edition':'Edition',
                                   'edition_id':'Edition_ID','sport':'Sport','sport_url':'Sport_URL',
                                   'result_location':'Result_Location','result_participants':'Result_Participants',
                                   'result_format':'Result_Format','result_detail':'Result_Detail',
                                   'result_description':'Result_Description','start_date':'Start_Date',
                                   'end_date':'End_Date','time':'Time'})
results.columns
```

```python
# Splitting Edition into list
results['column'] = results["Edition"].str.split(' ')

results.head()
```

```python
# Making new DataFrame to split 
results_models = pd.DataFrame(results['column'].tolist()).fillna('').add_prefix('model_')
results_models = results_models.drop(['model_2','model_3'], axis = 1)
results_models.head()
```

```python
# Concatinating DFs
new_results = pd.concat([results,results_models], axis = 1)

new_results.head()
```

```python
# Dropping unneeded columns and renaming new columns
new_results = new_results.drop(['column','Edition','Edition_ID','Sport_URL','Result_Format','Result_Description','Result_Participants'], axis = 1)
new_results = new_results.rename(columns = {'model_0':'Year','model_1':'Season'})

new_results.head()
```

```python
# Looking for na in Start_Date due to being unable to convert datatype

new_results.loc[(new_results['Start_Date'] == 'na')]
```

```python
# Changing Start_Date, End_Date, and Year columns datatypes
new_results["Start_Date"] = pd.to_datetime(new_results['Start_Date'], errors = 'coerce')

new_results['End_Date'] = pd.to_datetime(new_results['End_Date'], errors = 'coerce')

new_results['Year'] = pd.to_numeric(new_results['Year'], errors = 'coerce')

new_results.dtypes
```

```python
# Making csv file (initial)

new_results.to_csv('clean_results.csv', index = False, header = True)
```

```python
# Reading in csv
games = pd.read_csv('../Resources/Olympics_Games.csv')

games.head()
```

```python
# Checking columns for unneeded characters

games.columns
```

```python
games.loc[(games['start_date'] == 'na')]                 
```

```python
# Changing column names for parity

games = games.rename(columns = {'edition':'Edition','edition_id':'Edition_ID','edition_url':'Edition_URL',
                                'season':'Season', 'year':'Year','city':'City','country_flag_url':'Country_Flag_URL',
                               'country_noc':'Country_Code','start_date':'Start_Date','end_date':'End_Date',
                               'competition_start_date':'Competition_Start_Date',
                                'competition_end_date':'Competition_End_Date'})

games.columns
```

```python
# Removing Olympics that were not held

games = games.drop([5, 11, 12, 32, 33, 34, 39, 40, 61])

games.loc[(games['Start_Date'] == 'na')]
```

```python
# Dropping unneeded columns

games = games.drop(['Edition','Edition_URL','Country_Flag_URL', 'isHeld'],
                   axis = 1, inplace = False)

games.head()
```

```python
games.dtypes
```

```python
# Converting columns datatypes
games["Competition_Start_Date"] = pd.to_datetime(games['Competition_Start_Date'], errors = 'coerce')

games['Competition_End_Date'] = pd.to_datetime(games['Competition_End_Date'], errors = 'coerce')

games['Start_Date'] = pd.to_datetime(games['Start_Date'], errors = 'coerce')

games['End_Date'] = pd.to_datetime(games['End_Date'], errors = 'coerce')

games.dtypes
```

```python
# Making csv file (initial)

games.to_csv('clean_games.csv',index = False,header = True)
```
