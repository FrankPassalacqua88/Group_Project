```python
import pandas as pd
```

```python
# Reading in csv
gdp = pd.read_csv('../Resources/Original/gdp_1960_2021.csv')

gdp.head()
```

```python
# Renaming columns for parity
gdp = gdp.rename(columns = {'Country Name':'Country_Name',
                            'Country Code':'Country_Code',
                            'Indicator Name':'Indicator_Name',
                           'Indicator Code':'Indicator_Code'})
gdp = gdp.drop(['Indicator_Name','Indicator_Code'], axis = 1, inplace = False)

gdp.columns
```

```python
# Changing all year columns to row values under Year column

gdp2 = pd.melt(gdp,id_vars =['Country_Name','Country_Code'], var_name = 'Year', value_name = 'Value')

gdp2.head(10)
```

```python
# Changing Year column datatype
gdp2['Year'] = pd.to_numeric(gdp2['Year'], errors = 'coerce')
gdp2.dtypes
```

```python
# Reading in csv
medals = pd.read_csv('Athletes/clean_medal_tally.csv')

medals.head()
```

```python
# Merging the two DFs together
merge_df = pd.merge(gdp2,medals, on = ['Country_Code','Year'])

merge_df
```

```python
# Dropping, renaming columns
merge_df = merge_df.drop(['Country_Name_y'], axis = 1)

merge_df = merge_df.rename(columns = {'Country_Name_x':'Country_Name', 'Value':'GDP'})
```

```python
# Making DF of only Summer Olympics
gdp_summer = merge_df.loc[(merge_df['Season']== 'Summer')]

gdp_summer = gdp_summer.drop(['Season'], axis = 1)

gdp_summer
```

```python
# Making DF of only Winter Olympics
gdp_winter = merge_df.loc[(merge_df['Season'] == 'Winter')]

gdp_winter = gdp_winter.drop(['Season'], axis = 1)

gdp_winter
```

```python
# Making csv files of DFs
merge_df.to_csv('gpd_medals_total.csv', index = False)

gdp_summer.to_csv('gdp_medals_summer.csv', index = False)

gdp_winter.to_csv('gdp_medals_winter.csv', index = False)
```

```python
# Reading in csv
pop = pd.read_csv('../Resources/clean_population.csv')

pop.head()
```

```python
# Dropping column and converting year columns into values in new column Year
pop = pop.drop(['Unnamed: 0'], axis = 1)

pop2 = pd.melt(pop,id_vars =['Country_Name','Country_Code'], var_name = 'Year', value_name = 'Value')

pop2
```

```python
# Converting Year column datatype
pop2['Year'] = pd.to_numeric(pop2['Year'], errors = 'coerce')
pop2.dtypes
```

```python
# Merging DFs together
pop_merge_df = pd.merge(pop2,medals, on = ['Country_Code','Year'])

pop_merge_df
```

```python
# Dropping, renaming columns
pop_merge_df = pop_merge_df.drop(['Country_Name_y'], axis = 1)

pop_merge_df = pop_merge_df.rename(columns = {'Country_Name_x':'Country_Name', 'Value':'Population'})

pop_merge_df
```

```python
# Making DF of only Summer Olympics
pop_summer = pop_merge_df.loc[(pop_merge_df['Season'] == 'Summer')]

pop_summer = pop_summer.drop(['Season'], axis = 1)

pop_summer
```

```python
# Making DF of only Winter Olympics
pop_winter = pop_merge_df.loc[(pop_merge_df['Season'] == 'Winter')]

pop_winter = pop_winter.drop(['Season'], axis = 1)

pop_winter
```

```python
# Making csv files of DFs
pop_merge_df.to_csv('pop_medals_total.csv', index = False)

pop_summer.to_csv('pop_medals_summer.csv', index = False)

pop_winter.to_csv('pop_medals_winter.csv', index = False)
```
